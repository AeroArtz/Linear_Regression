```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/real-estate-price-prediction/Real estate.csv')
```

```python
df.head()
```

```python
X = df.drop('Y house price of unit area',axis=1)
Y = df['Y house price of unit area']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
x_train,x_test,y_train,y_test = x_train/100,x_test/100,y_train/100,y_test/100
class LinearRegression:
    def __init__(self,alpha,epochs): #initializing hyperparameters
        self.alpha = alpha  #learning rate
        self.epochs = epochs  #no. of iterations
    def fit(self,X,Y): #inializing wieghts and biases
        self.X = X
        self.Y = Y
        self.m,self.n = X.shape
        self.b = 0
        self.w = np.zeros(self.n)
        for i in range(self.epochs):    #updating parameters
            self.gradient_descent()
        return self
        
    def gradient_descent(self):     
        y_pred = self.pred(self.X)
        
        #derivatives of loss function with respect to parametes
        dw = 2 * self.X.T.dot(y_pred-self.Y) * 1/self.m
        db = 2* np.sum(y_pred-self.Y)/self.m
        
        #updating parameters
        self.w = self.w - (self.alpha*dw)
        self.b = self.b - (self.alpha*db)
        
        return self
    
    def accuracy(self,X,Y):
        prediction = self.pred(X)
        print(f'predicted values are : {prediction}')
        print(f'actual values: {Y}')
        
    def mse(self,X_test,Y_test):
        y_pred = self.pred(X_test)
        return 1/self.m * np.sum(y_pred-Y_test)
        
        
        
    def pred(self,x):
        y = x.dot(self.w)+self.b
        return y
        
```

```python
model = LinearRegression(0.001,3000)
model.fit(x_train,y_train)

```

```python
print(f'loss: {model.mse(x_test,y_test)}')
```
